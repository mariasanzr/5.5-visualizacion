# Data cleansing

```python
# Import of libraries

import pandas as pd
import numpy as np
```

```python
# Import and display of the CSV

st = pd.read_csv('../data/streaming_viewership_data.csv')
st.head()
```

```python
# Info of the data frame

st.info()
```

```python
# Checking the range of ages

unique = st['Age'].unique()
unique.sort()
unique
```

```python
# Adding a new column to divide age into ranges

conditions = [
    (st['Age'] >= 10) & (st['Age'] < 30),
    (st['Age'] >= 30) & (st['Age'] < 50),
    (st['Age'] >= 50) & (st['Age'] <= 70)]

values = ['young', 'middle_age', 'senior']

st['Age_Range'] = np.select(conditions, values)
```

```python
# Checking the results

st.head()
```

```python
# Export of the data frame 

st.to_csv('streaming_viewership.csv', index = False)
```
